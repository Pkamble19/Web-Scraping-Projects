```python
import pandas as pd
from bs4 import BeautifulSoup
import requests ,openpyxl
exel=openpyxl.Workbook()
sheet=exel.active
sheet.title='Boat Earphones'
sheet.append(['Name','Orignal Price','Discount','Offer Price','Rating Out Of 5','Total reviews'])
try:
    url='https://www.boat-lifestyle.com/collections/bluetooth-wireless-headphones?sort_by=best-selling&filter.v.availability=1&filter.v.price.gte=&filter.v.price.lte='
    source=requests.get(url)
    soup=BeautifulSoup(source.text,'html.parser')
    earphones=soup.find_all('product-item',class_='product-item')
    for earphone in earphones:
        name=earphone.find('a',class_='product-item-meta__title tile-title').text
        price=earphone.find('span',class_="price price--highlight product-card-price").text.split('â‚¹')[1]
        rating=earphone.find('div',class_="rating").text.split(' ')[0]
        total_rev=earphone.find('span',class_="rating__caption").text
        og_price=earphone.find('span',class_="price price--compare line-through").text.split(" ")[2]
        discount=earphone.find('p',class_="custom-save-price").text.split('(')[1].split(')')[0]
        sheet.append([name,og_price,discount,price,rating,total_rev])
except Exception as e:
    print(e)
exel.save('Boat Earphones.xlsx')
```
